```python
import folium #map rendering library
import requests #getting website page library
import pandas as pd #data analysis library
import numpy as np #array and vector handle library
from pandas.io.json import json_normalize #convert JSON file to pandas datafrae
from geopy.geocoders import Nominatim #get latitude and longitude using address
from bs4 import BeautifulSoup #turn webpage to readabe data
```

```python
#fetch data from website
source = requests.get('https://worldpostalcode.com/malaysia/selangor').text
soup = BeautifulSoup(source, 'html.parser')
```

```python
#convert data feteched into pandas dataframe
li_area = []
li_code = []

for row in soup.find_all('div', {'class':'leftc'}):
    li_area.append(row.text)
    
for row in soup.find_all('div', {'class':'rightc'}):
    li_code.append(row.text)

df_raw = pd.DataFrame({'Area':li_area,'Postcode':li_code})
df_raw.drop([7,19,27], inplace=True)
df_raw.replace('Kuala Kubu Baru','Kuala Kubu Bharu', inplace=True)
df_raw.replace('Telok Panglima Garang','Teluk Panglima Garang', inplace=True)
df_raw[:10]
```

```python
#turn value in 'Postcode' column into single postcode for each row 
df_raw_coded = pd.concat([pd.Series(row['Area'], row['Postcode'].split()) for i, row in df_raw.iterrows()])
df_raw_coded = pd.DataFrame(df_raw_coded)
df_raw_coded.reset_index(inplace=True)
df_raw_coded = df_raw_coded.rename(columns={'index':'Postcode', 0:'Area'})
df_raw_coded = df_raw_coded[['Area','Postcode']]
df_raw_coded[:10]
```

```python
#making sure the size of table and count the postcode unique value
print(df_raw_coded.shape[0])
print(len(df_raw_coded['Postcode'].unique()))
```

```python
#getting latitude and longitude for Selangor, Malaysia
address = 'Selangor, Malaysia'

geolocator = Nominatim(user_agent='my-application')
location = geolocator.geocode(address)
lat_selangor = location.latitude
lng_selangor = location.longitude
```

```python
#getting latitude and longitude for whole area in Selangor, Malaysia
li_postcode_raw = []
li_lat_raw = []
li_lng_raw = []

for row in df_raw_coded['Area']:
    address = row + ', Selangor, Malaysia'
        
    geolocator = Nominatim(user_agent='my-application')
    location = geolocator.geocode(address)
        
    if location.raw['display_name'][-15:-10] == 'angor':
        li_postcode_raw.append('0')
    else:
        li_postcode_raw.append(location.raw['display_name'][-15:-10])
        
    li_lat_raw.append(location.latitude)
    li_lng_raw.append(location.longitude)
```

```python
#making sure the number postcode, latitude and longitude is the same
print(len(li_postcode_raw), len(li_lat_raw), len(li_lng_raw))
```

```python
#put the postcode, lattude and longitude in dataframe
df_postcode_raw = pd.DataFrame({'Postcode_':li_postcode_raw,'Latitude':li_lat_raw,'Longitude':li_lng_raw})
df_postcode_raw[:10]
```

```python
#combine dataframe from website and from Nominatim
df_selangor = pd.concat([df_raw_coded,df_postcode_raw], axis=1, sort=False)
df_selangor[:10]
```

```python
#drop redudant data column after checking each postcode is same except for '0'
df_selangor.drop(['Postcode_'], axis=1, inplace=True)
df_selangor[:10]
```

```python
#create map to indicate each areas in Selangoor,Malaysia
map_selangor= folium.Map(location=[lat_selangor, lng_selangor], zoom_start=9)

for area, postcode, lat, lng in zip(df_selangor['Area'], df_selangor['Postcode'], df_selangor['Latitude'], df_selangor['Longitude']):
    label = '{0}, {1}'.format(postcode, area)
    popup = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng],
                        radius=5,
                        popup=popup,
                        color='blue',
                        fill=True,
                        fill_color='blue',
                        fill_opacity=0.7
                       ).add_to(map_selangor)  
    
map_selangor
```

```python
#list of fuction that used in this project
def getCategoryName(row):
    if len(row['Venue_Category']) == 0:
        return None
    else:
        return row['Venue_Category'][0]['name']

def getNearbyVenues(names, latitudes, longitudes):
    li_area_name = []
    li_area_lat = []
    li_area_lng = []
    li_venue_name = []
    li_venue_cat = []
    li_venue_lat = []
    li_venue_lng = []
    df_venues_list = pd.DataFrame()
    
    for name, lat, lng in zip(names, latitudes, longitudes):
        url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            client_id, 
            client_secret, 
            ver, 
            lat, 
            lng, 
            radius, 
            limit)
        
        result = requests.get(url).json()['response']
        
        for row in result['venues']:
            li_area_name.append(name)
            li_area_lat.append(lat)
            li_area_lng.append(lng)
            li_venue_name.append(row['name'])
            li_venue_cat.append(row['categories'])
            li_venue_lat.append(row['location']['lat'])
            li_venue_lng.append(row['location']['lng'])
    
        df_venues = pd.DataFrame({'Area':li_area_name,
                                  'Area_Latitude':li_area_lat,
                                  'Area_Longitude':li_area_lng,
                                  'Venue':li_venue_name,
                                  'Venue_Category':li_venue_cat,
                                  'Venue_Latitude':li_venue_lat,
                                  'Venue_Longitude':li_venue_lng
                                 }) 
        
    df_venues_list = pd.concat([df_venues_list, df_venues], sort=False)
    
    return df_venues_list

def getTypeCount(dataframe, name):
    li_entertainment = list(df_entertainment['Arts_Entertainment'])
    li_uni = list(df_uni['College_University'])
    li_event = list(df_event['Event'])
    li_food = list(df_food['Food'])
    li_night = list(df_night['Nightlife'])
    li_recreation = list(df_recreation['Outdoors_Recreation'])
    li_other = list(df_other['Professional_Other'])
    li_residence = list(df_residence['Residence'])
    li_service = list(df_service['Shop_Service'])
    li_travel = list(df_travel['Travel_Transport'])

    df = pd.DataFrame([
        ['Arts_Entertainment', dataframe[dataframe['Venue_Category'].isin(li_entertainment)]['Venue_Count'].sum()], 
        ['College_University', dataframe[dataframe['Venue_Category'].isin(li_uni)]['Venue_Count'].sum()],
        ['Event', dataframe[dataframe['Venue_Category'].isin(li_event)]['Venue_Count'].sum()],
        ['Food', dataframe[dataframe['Venue_Category'].isin(li_food)]['Venue_Count'].sum()],
        ['Nightlife', dataframe[dataframe['Venue_Category'].isin(li_night)]['Venue_Count'].sum()],
        ['Outdoors_Recreation', dataframe[dataframe['Venue_Category'].isin(li_recreation)]['Venue_Count'].sum()],
        ['Professional_Other', dataframe[dataframe['Venue_Category'].isin(li_other)]['Venue_Count'].sum()],
        ['Residence', dataframe[dataframe['Venue_Category'].isin(li_residence)]['Venue_Count'].sum()],
        ['Shop_Service', dataframe[dataframe['Venue_Category'].isin(li_service)]['Venue_Count'].sum()],
        ['Travel_Transport', dataframe[dataframe['Venue_Category'].isin(li_travel)]['Venue_Count'].sum()],
    ], 
        columns=[name + '_Category', name + '_Count'])
    
    return df
```
